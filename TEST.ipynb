```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Define stocks
food_stock = patches.Rectangle((0.2, 0.5), 0.2, 0.2, linewidth=2, edgecolor='b', facecolor='none')
waste_stock = patches.Rectangle((0.6, 0.5), 0.2, 0.2, linewidth=2, edgecolor='b', facecolor='none')

# Define flows
production_arrow = patches.Arrow(0.05, 0.5, 0.1, 0, linewidth=2, edgecolor='g')
consumption_arrow = patches.Arrow(0.45, 0.6, 0.1, 0, linewidth=2, edgecolor='r')
waste_arrow = patches.Arrow(0.75, 0.5, 0.1, 0, linewidth=2, edgecolor='r')

# Add stocks and flows to the axis
ax.add_patch(food_stock)
ax.add_patch(waste_stock)
ax.add_patch(production_arrow)
ax.add_patch(consumption_arrow)
ax.add_patch(waste_arrow)

# Add labels to stocks and flows
ax.text(0.3, 0.7, 'Food Stock', fontsize=12, color='b')
ax.text(0.7, 0.7, 'Waste Stock', fontsize=12, color='b')
ax.text(0.03, 0.52, 'Production', fontsize=10, color='g')
ax.text(0.43, 0.62, 'Consumption', fontsize=10, color='r')
ax.text(0.83, 0.52, 'Waste', fontsize=10, color='r')

# Set axis limits and labels
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

# Add a title
plt.title('Stock-and-Flow Diagram: Food Production and Waste', fontsize=14)

# Display the diagram
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Define nodes and add them to the graph
nodes = [
    ('Food Stock', 'Food Stock'),
    ('Waste Stock', 'Waste Stock'),
    ('Production', 'Production'),
    ('Consumption', 'Consumption'),
    ('Waste', 'Waste'),
]

G.add_nodes_from(nodes)

# Define edges and add them to the graph
edges = [
    ('Production', 'Food Stock'),
    ('Food Stock', 'Consumption'),
    ('Consumption', 'Waste'),
    ('Waste', 'Waste Stock'),
]

G.add_edges_from(edges)

# Set positions for nodes to create a more organized layout
pos = {
    'Food Stock': (0, 1),
    'Waste Stock': (2, 1),
    'Production': (0, 0),
    'Consumption': (2, 0),
    'Waste': (1, -1),
}

# Draw the graph with improved layout and styling
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_size=1000, node_color='lightblue', font_size=10, font_weight='bold', arrowsize=20, edge_color='gray', width=1)
plt.title("Stock-and-Flow Diagram: Food Production and Waste", fontsize=14)
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Define nodes and add them to the graph
nodes = [
    ('Food Stock', {'pos': (0, 0)}),
    ('Waste Stock', {'pos': (4, 0)}),
    ('Production', {'pos': (1, 2)}),
    ('Consumption', {'pos': (3, 2)}),
    ('Waste', {'pos': (2, -2)}),
]

G.add_nodes_from(nodes)

# Define edges and add them to the graph
edges = [
    ('Production', 'Food Stock'),
    ('Food Stock', 'Consumption'),
    ('Consumption', 'Waste'),
    ('Waste', 'Waste Stock'),
]

G.add_edges_from(edges)

# Get node positions
node_positions = nx.get_node_attributes(G, 'pos')

# Draw the graph with specified positions and styling
plt.figure(figsize=(8, 6))
nx.draw(G, pos=node_positions, with_labels=True, node_size=1000, node_color='lightblue', font_size=10, font_weight='bold', arrowsize=20, edge_color='gray', width=1)
plt.title("Stock-and-Flow Diagram: Food Production and Waste", fontsize=14)
plt.axis('off')
plt.tight_layout()
plt.show()
```
