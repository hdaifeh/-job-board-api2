```python
class Node:
  def __init__(self, name):
    self.name = name
    self.children = []

  def add_child(self, child):
    self.children.append(child)

  def get_name(self):
    return self.name

  def get_children(self):
    return self.children


def print_diagram(node, prefix):
  print(prefix + "|__ " + node.get_name())
  for child in node.get_children():
    print_diagram(child, prefix + "| ")


if __name__ == "__main__":

  # Root node
  root = Node("Food Waste Reduction")

  # First level nodes
  generation = Node("Waste Generation")
  collection = Node("Waste Collection")
  treatment = Node("Waste Treatment")
  disposal = Node("Waste Disposal")

  # Second level nodes
  prediction = Node("Prediction Models")
  efficiency = Node("Collection Efficiency")
  effectiveness = Node("Treatment Effectiveness")
  landfill = Node("Landfill Capacity")

  # Third level nodes
  solid_waste = Node("Municipal Solid Waste")
  components = Node("Components (e.g., food waste, green waste)")
  infrastructure = Node("Collection Infrastructure")
  treatment_methods = Node("Treatment Methods (e.g., composting, anaerobic digestion)")

  # Connect nodes
  root.add_child(generation)
  root.add_child(collection)
  root.add_child(treatment)
  root.add_child(disposal)

  generation.add_child(prediction)
  collection.add_child(efficiency)
  treatment.add_child(effectiveness)
  disposal.add_child(landfill)

  prediction.add_child(solid_waste)
  prediction.add_child(components)
  efficiency.add_child(infrastructure)
  effectiveness.add_child(treatment_methods)

  # Print diagram
  print_diagram(root, "")
```

```python
# Food waste reduction diagram

class Node:
  def __init__(self, name):
    self.name = name
    self.children = []

  def add_child(self, child):
    self.children.append(child)

def print_diagram(node, level=0):
  print("  " * level + node.name)
  for child in node.children:
    print_diagram(child, level+1)

if __name__ == "__main__":

  root = Node("Food Waste Reduction")

  # Waste Generation
  generation = Node("Waste Generation")
  root.add_child(generation)

  prediction = Node("Prediction Models")
  generation.add_child(prediction)

  msw = Node("Municipal Solid Waste")
  components = Node("Components (e.g. food, green)")
  prediction.add_child(msw)
  prediction.add_child(components)

  # Waste Collection
  collection = Node("Waste Collection")
  root.add_child(collection)

  efficiency = Node("Collection Efficiency")
  collection.add_child(efficiency)

  infrastructure = Node("Infrastructure")
  efficiency.add_child(infrastructure)

  # Waste Treatment
  treatment = Node("Waste Treatment")
  root.add_child(treatment)

  methods = Node("Treatment Methods")
  treatment.add_child(methods)

  compost = Node("Composting")
  ad = Node("Anaerobic Digestion")
  methods.add_child(compost)
  methods.add_child(ad)

  # Waste Disposal
  disposal = Node("Waste Disposal")
  root.add_child(disposal)

  landfill = Node("Landfill Capacity")
  disposal.add_child(landfill)

  # Print diagram
  print_diagram(root)
```
